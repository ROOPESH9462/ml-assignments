```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

this is the content which are present in the file which i selected for my  EDA work

```python
df = pd.read_csv('/content/sleep_health_lifestyle_dataset.csv')
df
```

it is the describe of the data

```python
df.describe().T
```

it is the graph that tells us how many people can have the sleeping disorder and another hand tells how many  people can not have the sleeping disorder

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Sleep Disorder'], palette="Set2")
plt.title("Distribution of Sleep Disorder", fontsize=16)
plt.xlabel("Sleep Disorder", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks([0, 1], labels=['No Disorder', 'Disorder'], fontsize=10)
plt.show()
```

in this repersention including the apnea details with comparision of sleep disorder

```python
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Sleep Disorder'].fillna('NaN'), palette="Set2")
plt.title("Distribution of Sleep Disorder (Including NaN)", fontsize=16)
plt.xlabel("Sleep Disorder", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

all disorders comparing's

```python
sleep_disorder_counts = df.groupby(['Occupation', 'Sleep Disorder']).size().unstack()

sleep_disorder_counts.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')
plt.title('Sleep Disorders by Occupation')
plt.legend(title='Sleep Disorder')
plt.show()
```

sleep duration of male and femail comparision

```python
sns.boxplot(x="Gender", y="Sleep Duration (hours)", data=df, hue="Gender", palette="coolwarm")
plt.title("Sleep Duration Comparison by Gender")
plt.xlabel("Gender")
plt.ylabel("Sleep Duration (hours)")
plt.show()
```
